<p>Optimasi parameter yang dilakukan hanya untuk beberapa parameter saja, dikarenakan keterbatasan waktu dalam pengerjaan.</p>


```python
# import the necessary packages
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.datasets import load_iris
from sklearn.datasets import load_wine
import numpy as np
```

```python
dataset = load_iris()
(trainX, testX, trainY, testY) = train_test_split(dataset.data, dataset.target, random_state=3, test_size=0.25)
```

```python
dataset1 = load_wine()
(trainX1, testX1, trainY1, testY1) = train_test_split(dataset1.data, dataset1.target, random_state=3, test_size=0.25)
```

## 1. Optimasi Parameter K-NN (Case: Iris)

```python
def model (n_neighbors):
    best_accuracy = 0
    for n_neighbors in range(1,n_neighbors):
        parameter=(n_neighbors)
        model=KNeighborsClassifier(n_neighbors=n_neighbors)
        model.fit(trainX, trainY)
        print("[INFO] evaluating...")
        print("n_neighbors=",n_neighbors)
        predictions = model.predict(testX)
        accuracy=classification_report(testY, predictions,target_names=dataset.target_names,output_dict=True)['accuracy']
        print('accuracy =',accuracy)
        if accuracy > best_accuracy:
            best_accuracy, best_parameter = accuracy, parameter
    print('Best K-NN Parameter(n_neighbors)=%s Accuracy=%f' % (best_parameter, best_accuracy))

model(50)
```

## 2. Optimasi Parameter Naive Bayes (Case: Iris)

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,ComplementNB,BernoulliNB
model=list([GaussianNB(),MultinomialNB(),ComplementNB(),BernoulliNB()])
max_alpha=1.0
best_accuracy=0
# train the NaÃ¯ve Bayes model
for model in model:
    for alpha in np.arange(0,max_alpha,0.1):
        parameter=(model,alpha)
        model.fit(trainX, trainY)
        # make predictions on our data and show a classification report
        print("[INFO] evaluating...")
        print("Model Naive Bayes =",model, "alpha=",alpha)
        predictions = model.predict(testX)
        accuracy=classification_report(testY, predictions,target_names=dataset.target_names,output_dict=True)['accuracy']
        print('accuracy=',accuracy)
        if accuracy > best_accuracy:
            best_accuracy, best_parameter = accuracy, parameter
    print('Best Naive Bayes Parameter(n_neighbors)=%s Accuracy=%f' % (best_parameter, best_accuracy))
```

## 3. Optimasi Parameter Logistic Regression (Case: Iris)

```python
def model_LogisticRegression (Cs,solvers):
    best_accuracy = 0
    for C in range (1,Cs):
        for solver in solvers:
            parameter=(C,solver)
            model=LogisticRegression(C=C,solver=solver, multi_class="auto")
            model.fit(trainX, trainY)
            print("[INFO] evaluating...")
            print("LogisticRegression, C=",C,"solver= ",solver,'multi_class=auto')
            predictions = model.predict(testX)
            #print(classification_report(testY, predictions,target_names=dataset.target_names))
            accuracy=classification_report(testY, predictions,target_names=dataset.target_names,output_dict=True)['accuracy']
            print('accuracy=',accuracy)
            if accuracy > best_accuracy:
                best_accuracy, best_parameter = accuracy, parameter
    print('Best Logistic Regression Parameter(C,solver)=%s Accuracy=%f' % (best_parameter, best_accuracy))

solver=list(['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']) 
Cs=50
model_LogisticRegression(Cs,solver)
```

## 4. Optimasi Parameter SVM (Case: Iris)

```python
def model (Cs,gammas):
    best_accuracy = 0
    for C in range (1,Cs):
        for gamma in np.arange (0.01,gammas,0.01):
            parameter=(C,gamma)
            model=SVC(kernel="rbf", gamma=gamma,C=C)
            model.fit(trainX, trainY)
            print("[INFO] evaluating...")
            print("SVC kernel  RBF, C=",C,"gamma= ",gamma)
            predictions = model.predict(testX)
            #print(classification_report(testY, predictions,target_names=dataset.target_names))
            accuracy=classification_report(testY, predictions,target_names=dataset.target_names,output_dict=True)['accuracy']
            print('accuracy=',accuracy)
            if accuracy > best_accuracy:
                best_accuracy, best_parameter = accuracy, parameter
    print('Best SVM Parameter(C,gamma)=%s Accuracy=%f' % (best_parameter, best_accuracy))
    
model(10,1)
```
